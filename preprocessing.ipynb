```python
df_count=pd.DataFrame()
for file in files:
    df=pd.read_csv(file,sep='\t',header=None,index_col=0)
    df.columns= [os.path.basename(file).split('_')[0]] #name the columns with the filename
    df=df.iloc[:-5,] #remove last 5 lines of every file 
    df_count=pd.merge(left=df_count,right=df,how='outer',left_index=True,right_index=True)
```

```python
df_count
```

```python
df_rpm=df
for column in df_rpm.columns:
    Total = df_rpm[column].sum()
    df_rpm[column]=(df_rpm[column]/Total)*1000000

df_rpm
```
